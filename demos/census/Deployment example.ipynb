```python
from verta import Client
from verta.utils import ModelAPI

VERTA_HOST = "https://cm.dev.verta.ai"

client = Client(VERTA_HOST)
```

```python
endpoint = client.get_or_create_endpoint("/census")
```

```python
project = client.set_project(name="Census Income S3")
run = project.expt_runs.top_k('metrics.val_acc', 1)[0]

endpoint.update(run, wait=True)
```

```python
deployed_model = endpoint.get_deployed_model()
```

```python
import time
import itertools
import pandas as pd

df_test = pd.read_csv("census-test.csv")
X_test = df_test.iloc[:,:-1]

for x in itertools.cycle(X_test.values.tolist()):
    print(deployed_model.predict([x]))
    time.sleep(.2)
```
